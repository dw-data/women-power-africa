Women who believe a wife is justified refusing sex with her husband (%)

This indicator measures the percentage of women aged 15-49 who believe that a wife is justified in refusing to have sex with her husband. The indicator includes breakdowns by reason for refusal, and by wealth quintile.

ID: WB_GS_SG_RSX
https://data360.worldbank.org/en/indicator/WB_GS_SG_RSX

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("WB_GS_SG_RSX.csv")
```

```python
df_africa = df[df['REF_AREA_LABEL'].isin(africa_countries)]
```

```python
df_africa['COMP_BREAKDOWN_2_LABEL'].value_counts()
```

```python
df_africa_selected = df_africa.groupby(['REF_AREA_LABEL', 'TIME_PERIOD'])['OBS_VALUE'].mean().reset_index().dropna()
```

```python
df_africa_selected['Area'] = 'refuse sex'
```

```python
df_africa_selected.to_csv("beliefRefuseSex.csv", index=False)
```
